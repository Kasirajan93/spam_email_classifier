```python
import pandas as pd
```

```python
dataset1 = pd.read_csv("X:/AI-SPAM-PHISHING-DETECTOR/data/processed/email_spam_cleaned.csv")

dataset2 = pd.read_csv("X:/AI-SPAM-PHISHING-DETECTOR/data/processed/personal_emails_cleaned.csv")

dataset3 = pd.read_csv("X:/AI-SPAM-PHISHING-DETECTOR/data/processed/spamassassin_cleaned.csv")
```

```python
print("Dataset 1 :", dataset1.shape)
print("Dataset 2 :", dataset2.shape)
print("Dataset 3 :", dataset3.shape)
```

```python
# Dataset 1
dataset1 = dataset1[["title", "text", "type"]]

# Dataset 2
dataset2 = dataset2[["title", "text", "type"]]

# Dataset 3
dataset3 = dataset3[["title", "text", "type"]]
```

```python
final_df = pd.concat(
    [
        dataset1,
        dataset2,
        dataset3
    ],
    ignore_index=True
)
```

```python
print(final_df["type"].value_counts())
```
